```python
import numpy as np
import pandas as pd
```

```python
games = pd.read_csv("vgchartz-2024.csv")
```

```python
games.sample(5)
```

```python
games.info()
```

```python
games.describe()
```

```python
games.dropna(subset=["total_sales"], inplace=True)
```

```python
games = games.drop("img", axis=1)
```

```python
games["release_date"] = pd.to_datetime(games["release_date"])
```

```python
games[["publisher","developer"]] = games[["publisher","developer"]].fillna("Unknown")
```

```python
games.info()
```

```python
games["year"] = games["release_date"].dt.year
```

```python
games["year"] = games["year"].astype("Int64")
```

```python
sales_cols = ["na_sales", "jp_sales", "pal_sales", "other_sales"]
games[sales_cols] = games[sales_cols].fillna(0)
```

```python
games["calculated_total"] = games[sales_cols].sum(axis=1)
```

```python
games["difference"] = (games["calculated_total"] - games["total_sales"]).round(2)
incorrect_rows = games[games["difference"].abs() > 0.01]
print("Number of inconsistent rows:", len(incorrect_rows))
```

```python
games[games["difference"].abs() > 0.01]
```

```python
games.describe()
```

```python
avg_sales_by_genre = games.groupby("genre")["total_sales"].mean().sort_values(ascending=False)
print(avg_sales_by_genre)
```

```python
top_genre = avg_sales_by_genre.idxmax()
top_sales = avg_sales_by_genre.max()
print(f"The genre with the highest average total sales is '{top_genre}' with {top_sales:.2f} million units sold on average.")
```

```python
minecraft_rows = games[games["title"].str.contains("Minecraft", case=False, na=False)]
minecraft_rows
```

```python
games = games.drop(["calculated_total", "difference"], axis=1)
```

```python
games.info()
```

```python
games.to_csv("vgchartz-2024_cleaned.csv", index=False)
```
